```python
import numpy as np 
from numpy.fft import fft,ifft,fftshift
import matplotlib.pyplot as plt
```

```python
tau = 1/2
freq = np.arange(-4,4)
freq
```

```python
f_tabs = np.exp((-2j*np.pi*freq*tau)/8)
f_tabs
```

```python
plt.plot(f_tabs.real)
plt.plot(f_tabs.imag)
plt.show()
```

```python
i_tabs = ifft(fftshift(f_tabs))
i_tabs
```

```python
shiftet= fftshift(f_tabs)
plt.plot(shiftet.real)
plt.plot(shiftet.imag)
plt.show()
```

```python
plt.plot(i_tabs.real)
plt.plot(i_tabs.imag)
plt.show()
```

# Delay as a Convolution System

```python
# Create the FIR
D = 3 # Delay value
h = [0]*D+[1]
h
```

```python
d = 1
k = np.linspace(-5,7,1000)
h_ideal = np.sinc(k-d)
plt.plot(k,h_ideal)
plt.show()
```

```python
n = 4
window_sample= np.arange(np.round(d-(n-1)/2),np.round(d+(n+1)/2))
window = []
for ws in window_sample:  
    idx = (np.abs(k-ws)).argmin()
    window.append(idx)
    
plt.plot(k,h_ideal)
plt.plot(window_sample,h_ideal[window],"*")
plt.show()
```
